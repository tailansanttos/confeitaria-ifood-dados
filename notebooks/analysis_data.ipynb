```python
import pandas as pd

df = pd.read_csv("../data/raw/relatorios.csv")
df.head()
```

```python
filtro = ~(df['PRODUTO LOGISTICO'] == "ENTREGA PARCEIRA IFOOD")
df[filtro]
```

```python
filtro_status = df["STATUS FINAL DO PEDIDO"] == "CONCLUIDO"
df = df[filtro_status]
df
```

```python
df["DATA E HORA DO PEDIDO"] = pd.to_datetime(df["DATA E HORA DO PEDIDO"], dayfirst=True)
df["DATA PEDIDO"] = df["DATA E HORA DO PEDIDO"].dt.date
df["HORA PEDIDO"] = df['DATA E HORA DO PEDIDO'].dt.time
df
```
